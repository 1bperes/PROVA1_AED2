```python
class No:
    def __init__(self, valor):
        self.valor = valor
        self.proximo = None

class ListaEncadeada:
    def __init__(self):
        self.inicio = None

    def inserir(self, valor):
        if not isinstance(valor, int):
            return
        novo_no = No(valor)
        if self.inicio is None:
            self.inicio = novo_no
        else:
            atual = self.inicio
            while atual.proximo:
                atual = atual.proximo
            atual.proximo = novo_no

    def contar(self):
        atual = self.inicio
        contador = 0
        while atual:
            contador += 1
            atual = atual.proximo
        return contador


lista = ListaEncadeada()
lista.inserir(15)
lista.inserir(20)
lista.inserir(25)

print(lista.contar())
```
